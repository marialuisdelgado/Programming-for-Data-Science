```python
# Import library
import pandas as pd

# Read dataset
df = pd.read_csv('HR_Attrition_Dataset.csv')
```

```python
# EDA: Null values
df.isnull().sum() > 0
```

```python
# EDA: Look for duplicates
duplicate_rows_df = df[df.duplicated()]
print(duplicate_rows_df.shape)
```

```python
# Distribution on Attrition
attrition_dist = df['Attrition'].value_counts()
attrition_dist
```

```python
# Enconding Attrition (Yes = 1, No = 0)
df['Attrition_numeric'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
df_encoded = df.copy()
label_encoders = {}

for column in df_encoded.select_dtypes(include=['object']).columns:
  le = LabelEncoder()
  df_encoded[column] = le.fit_transform(df_encoded[column])
  label_encoders[column] = le

# Define target variable
X = df_encoded.drop(columns=['Attrition', 'Attrition_numeric', 'EmployeeNumber'])
y = df_encoded['Attrition_numeric']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Results on features
feature_importances = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)

# Top 5 most important features
feature_importances.head(5)
```
